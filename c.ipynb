```python
df1 = pd.get_dummies(df[["Profession","Gender"]])
df1
```

```python
df.drop(["Profession","Gender"], inplace = True, axis=1)
```

```python
df = pd.concat([df1], axis = 1)
df
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(df)
```

```python
x
```

```python
wcss = []

for i in range(1, 21):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 21), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') 
plt.show()
```

```python
dec = PCA(2)
dec1 = dec.fit_transform(x)
```

```python
plt.scatter(dec1[:, 0], dec1[:, 1])
```

```python
wcss = []

for i in range(1, 10):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(dec1)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 10), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') 
plt.show()
```

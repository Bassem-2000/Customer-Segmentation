```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import plotly.graph_objs as go
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
# from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.svm import LinearSVC, SVC
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, BaggingRegressor, GradientBoostingRegressor, ExtraTreesRegressor
sns.set()
```

```python
df = pd.read_csv('Customers.csv')
df = df.drop(['CustomerID', 'Profession'], axis = 1)
df['Gender'] = np.where(df['Gender']=='Male', 1, 0)
df
```

```python
df = df.dropna()
```

```python
test_data = df[(df['Age']<=17) | (df['Age']>=91)]
test_data 
```

```python
x_train = df.drop('Age', axis = 1)
# x_train = pd.get_dummies(x_train)
x_train.isnull().sum()
```

```python
y_train = df['Age']
y_train
```

```python
x_train1, x_test1, y_train1, y_test1 = train_test_split(x_train, y_train, train_size=0.8, random_state=0)
x_train1
```

```python
model = SVC()
model.fit(x_train1, y_train1)
```

```python
model.score(x_test1, y_test1)
# y_pred_test1 = model.predict(x_test1)
# r2_score(y_test1, y_pred_test1)
```

```python
x_test = test_data.drop('Age', axis = 1)
x_test = pd.get_dummies(x_test)
x_test
```

```python
y_pred = model.predict(x_test)
test_data['Age'] = y_pred
test_data['Age'] = test_data['Age'].round().astype(int)
test_data
```

```python
df.drop(["Profession","Gender", 'CustomerID'], inplace = True, axis=1)
df.drop(df[df.Age <= 17].index, axis= 0, inplace=True)
df.drop(df[df.Age >= 91].index, axis= 0, inplace=True)
```

```python
df
```

```python
test_data.drop(['Profession', 'Gender', 'CustomerID'], axis=1, inplace=True)
test_data
```

```python
df = pd.concat([df, test_data], axis = 0).sort_index()
```

```python
df
```

```python
df.min()
```

```python
x = StandardScaler().fit_transform(df)
x.max()
```

```python
x.min()
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, random_state = 0)
    kmeans.fit(x[:, 0:2])
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') 
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, max_iter = 100, n_init="auto", random_state=0)
y_kmeans = kmeans.fit_predict(x[:, 0:2])
y_kmeans
```

```python
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(x[:, 0], x[:, 1], x[:, 2], c = y_kmeans, cmap = 'tab20b', s =50)
```

```python
Scene = dict(xaxis = dict(title  = 'Age -->'),yaxis = dict(title  = 'Spending Score--->'),zaxis = dict(title  = 'Annual Income-->'))

# model.labels_ is nothing but the predicted clusters i.e y_clusters
labels = kmeans.labels_
trace = go.Scatter3d(x=x[:, 0], y=x[:, 1], z=x[:, 2], mode='markers',marker=dict(color = labels, size= 10, line=dict(color= 'black',width = 10)))
layout = go.Layout(margin=dict(l=0,r=0),scene = Scene,height = 800,width = 800)
data = [trace]
fig = go.Figure(data = data, layout = layout)
fig.show()
```
